Operators in python

```python
#Arithmetic Operators
a= 5
b =2

#addition
print(f'Addition: {a+b}')

#Subtration
print(f'Subtration: {a-b}')

#Multiply
print(f'Multiply: {a*b}')

#Divide
print(f'Divide: {a/b}')

# Floor Division
print("Floor Division:", a // b) 

#Modulus
print("Modulus:", a % b) 

# Exponentiation
print("Exponentiation:", a ** b)
```

```python
#Comparision operators
a= 23
b= 44

print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

```python
#Logical operator
#Python Logical operators perform Logical AND, Logical OR and Logical NOT operations. 
# It is used to combine conditional statements.
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

```python
# Bitwise Operators
# Python Bitwise operators act on bits and perform bit-by-bit operations. These are used to operate on binary numbers.

# Bitwise Operators in Python are as follows:

# Bitwise NOT
# Bitwise Shift
# Bitwise AND
# Bitwise XOR
# Bitwise OR

a = 6
b = 4

print(a & b)
print(a | b)
print(~a)
print(a ^ b)
print(a >> 2)
print(a << 2)
```

```python
# Assignment Operators
# Python Assignment operators are used to assign values to the variables. 
# This operator is used to assign the value of the right side of the expression to the left side operand.

a = 5
b = a
print(b)
b += a
print(b)
b -= a
print(b)
b *= a
print(b)
b <<= a
print(b)
```

```python
#Identity Operators
# is and is not are the identity operators both are used to check if two values are located on the same part of the memory. 
# Two variables that are equal do not imply that they are identical. 

# is          True if the operands are identical 
# is not      True if the operands are not identical 

a=10
b = 5
c=a 
d=10 

print(a is not b)
print(a is c)
print(c is d)
```

```python
#Membership operators
# "in" and "not in" are the membership operators that are used to test whether a value or variable is in a sequence.

x = 10
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

```python
#Ternary operators
#Syntax :  [on_true] if [expression] else [on_false] 
#allows to perfrom condition operations in one line

a, b = 10, 5
min = a if a < b else b

print(min)
```

```python
#Precedence and Associativity of Operators

## Operator Precedence
### This is used in an expression with more than one operator with different precedence to determine which operation to perform first.
expr = 10 + 20 * 30
print(expr)
name = "Anjal"
age = 1

if age <=2 and name == "Anjal" or name == "John":
    print("Hello! Welcome.")
else:
    print("Good Bye!!")
    

# Operator Associativity
# If an expression contains two or more operators with the same precedence then Operator Associativity is used to determine. 
# It can either be Left to Right or from Right to Left.
print("\nOutput of associativity...........")
print(100 / 10 * 10)
print(5 - 2 + 3)
print(5 - (2 + 3))
print(2 ** 3 ** 2)
```

#Data types in python
There are different types of data in python which are categorised as follows 
>Numeric --- Integer, Float , Complex Number
>Sequence --- String, List , Tuples
>Set --- set, frozenset
>Dictionary(Mapping)
>Boolean


```python
#Numeric Data types

a = 10
b=10.5
c = 10 + 11j

print(f"a is {type(a)} data types")
print(f"a is {type(b)} data types")
print(f"a is {type(c)} data types")
```

```python
#Sequence data types
##String
s = "Hello my name is anjal"

print(type(s))

print(s[1])
print(s[6])
print(s[-1])
```

```python
#Sequence data types
##List

a = [1,3,4,6,8]

print(a)

#ascessing list items

print(a[1])

print("Accessing element using negative indexing")
print(a[-1])
print(a[-3])

print(a[1:4])   # elements from index 1 to 3

#diferent list operations
a.append(10)
print("After appen 10",a)

a.insert(1,9)
print("After insert 9 0n 1",a)

a.extend([12,13,14])
print("After extend",a)



#revoing items from list
a.remove(1)
print("After removing 1",a)

a.pop(3)
print("After popping :",a)

del a[4]
print("After deleting index 4",a)

a.clear()
print("After clear():", a)
```

```python
#Sequence data types
##Tuples
#The only difference between a tuple and a list is that tuples are immutable. 
# Tuples cannot be modified after it is created.
a = (1,2,3,4,5)

print(type(a))



# access tuple items
print(a[0])
print(a[-1])
print(a[-3])
```

```python
#Boolean data types
print(type(True))
print(type(False))
```

```python
#Set data types
#Set is an unordered collection of data types that is iterable, mutable, and has no duplicate elements
set1 = set("Heyanjal")

set2= set(["Anjal","Hey","Anjal"])

print(set1)
print(set2)

#To access list items we can not use index cox it is un orderd so we use loop

for i in set1:
    print(i,end="-")

#Checking if items is in set or not
print("\n")
print("Anjal" in set2)
```

```python
#Dictionary Data Type
#Dictionary is like a mapping values
#it has key and values set

d = {1: 'Hey', 'name': 'Anjal', 3: '?'}

# Accessing an element using key
print(d['name'])

# Accessing a element using get
print(d.get(3))
```
